```python
def detect_faces(cascade, test_image, scaleFactor = 1.5):
    # create a copy of the image to prevent any changes to the original one.
    image_copy = test_image.copy()

    #convert the test image to gray scale as opencv face detector expects gray images
    gray_image = cv2.cvtColor(image_copy, cv2.COLOR_BGR2GRAY)

    # Applying the haar classifier to detect faces
    faces_rect = cascade.detectMultiScale(gray_image, scaleFactor=scaleFactor, minNeighbors=3)
    print("No. of faces found : " , len(faces_rect))
    for (x, y, w, h) in faces_rect:
        cv2.rectangle(image_copy, (x, y), (x+w, y+h), (0, 255, 0), 15)

    return image_copy
```

```python
#loading image
test_image2 = cv2.imread('team.jpg')

#call the function to detect faces
faces = detect_faces(haar_cascade_face, test_image2)

#convert to RGB and display image
plt.imshow(faces)
```

```python
def get_detected_faces(cascade, test_image, scaleFactor, minNeighbours):
    # create a copy of the image to prevent any changes to the original one.
    image_copy = test_image.copy()

    #convert the test image to gray scale as opencv face detector expects gray images
    gray_image = cv2.cvtColor(image_copy, cv2.COLOR_BGR2GRAY)

    # Applying the haar classifier to detect faces
    faces_rect = cascade.detectMultiScale(gray_image, scaleFactor=scaleFactor, minNeighbors= minNeighbours)
    print("No. of faces found : " , len(faces_rect))
    faces = []
    for (x, y, w, h) in faces_rect:
        cv2.rectangle(image_copy, (x, y), (x+w, y+h), (0, 255, 0), 15)
        face = gray_image[y:y+h, x:x+w]
        faces.append(face)

    return image_copy, faces
```

```python
test_image1 = io.imread('face.jpg')

#call the function to detect faces
full, faces = get_detected_faces(haar_cascade_face, test_image1, 1.1, 4)

#convert to RGB and display image
plt.imshow(faces[0])
# plt.imshow(full)
```
